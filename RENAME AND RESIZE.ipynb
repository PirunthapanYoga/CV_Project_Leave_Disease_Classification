```python
import os
import shutil
import random
import cv2
from PIL import Image 
import numpy as np

def agumentfile():
    srcfolder = "E:\\Project\\Dataset"
    destfolder = "E:\\Project_Modified\\Augmented\\Augmented_Full"
    for count,foldernames in enumerate(os.listdir(srcfolder)):
        path = os.path.join(destfolder,foldernames)
        os.mkdir(path)
        for count,filename in enumerate(os.listdir(srcfolder+"\\"+foldernames)):
            hashname=str(hash(filename.split(".")[0]))
            if(len(hashname.split("-"))>1):
                hashname=hashname.split("-")[1]
            dst=f"{hashname}{count}.jpg"
            src=f"{srcfolder}/{foldernames}/{filename}"
            dst=f"{destfolder}/{foldernames}/{dst}"
            
            #Real Image
            shutil.copyfile(src,dst)
            
            #Read Image
            image=cv2.imread(src)
            rgbImg=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            
            
            #Random Rotation
            (h, w) = image.shape[:2]
            (cX, cY) = (w // 2, h // 2)
            degree =random.randint(45,235)
            M = cv2.getRotationMatrix2D((cX, cY), degree, 1.0)
            rotated = cv2.warpAffine(rgbImg, M, (w, h))
            dst=f"{hashname}{count}r.jpg"
            dst=f"{destfolder}/{foldernames}/{dst}"
            im=Image.fromarray(rotated)
            im.save(dst)
            
            #Horizontal Flip 
            hFlip=cv2.flip(rgbImg, 1)
            dst=f"{hashname}{count}hf.jpg"
            dst=f"{destfolder}/{foldernames}/{dst}"
            im=Image.fromarray(hFlip)
            im.save(dst)
            
            #Vertical Flip 
            vFlip=cv2.flip(rgbImg, 0)
            dst=f"{hashname}{count}vf.jpg"
            dst=f"{destfolder}/{foldernames}/{dst}"
            im=Image.fromarray(vFlip)
            im.save(dst)
            
            #Brightness
            value = random.uniform(1, 1.5)
            hsv = cv2.cvtColor(rgbImg, cv2.COLOR_BGR2HSV)
            hsv = np.array(hsv, dtype = np.float64)
            hsv[:,:,1] = hsv[:,:,1]*value
            hsv[:,:,1][hsv[:,:,1]>255]  = 255
            hsv[:,:,2] = hsv[:,:,2]*value 
            hsv[:,:,2][hsv[:,:,2]>255]  = 255
            hsv = np.array(hsv, dtype = np.uint8)
            bimg = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
            dst=f"{hashname}{count}b.jpg"
            dst=f"{destfolder}/{foldernames}/{dst}"
            im=Image.fromarray(bimg)
            im.save(dst)
        
            #Channel Shift
            value = int(random.uniform(-25 , 75))
            img = rgbImg + value
            img[:,:,:][img[:,:,:]>255]  = 255
            img[:,:,:][img[:,:,:]<0]  = 0
            csimg = img.astype(np.uint8)
            dst=f"{hashname}{count}cs.jpg"
            dst=f"{destfolder}/{foldernames}/{dst}"
            im=Image.fromarray(csimg)
            im.save(dst)
            
```

```python
agumentfile()
```
